```python
import numpy as np
from numpy import arange
from scipy.optimize import curve_fit
from matplotlib import pyplot
def objective(x, m, b):
    return m*x+b

dist = np.array([147.4, 147.4, 126.7, 126.7, 103.2,103.2, 85.3, 85.3, 84.52, 84.52, 64.4, 64.4])
desp = np.array([56   ,  46  ,  45  ,  54  , 34   ,  66 , 81  ,  81 , 66   ,  80 ,  106, 107 ]) 
x, y = desp,dist
popt, _ = curve_fit(objective, x, y)
m,b = popt
pyplot.scatter(x, y)
x_line = arange(20,100, 1)
y_line = objective(x_line, m, b)
pyplot.plot(x_line, y_line, '--', color='red')
pyplot.ylabel('Distancia (cm)')
pyplot.xlabel('Desplazamiento (px)')
pyplot.show()
```
